### Loading data

**culled_pdb_rcc.csv** estos son los rccs del training set de proteins nativas.

**nativfoldedprots_rcc.csv** estos son los rccs de proteínas nativas que no están incluidas en culled_pdb_rcc.csv.

**decoys_4states_rcc.csv** y **decoys_loops_rcc.csv (1vfaA)** son los rccs de los decoys; el 4state son los modelos mas obviamente malos y el loops 


```python
import numpy as np
X_native_train_set = np.loadtxt("culled_pdb_rcc.csv", dtype="float", delimiter=",")
X_native_test_set = np.loadtxt("nativfoldedprots_rcc.csv", dtype="float", delimiter=",")
X_decoy_set = np.loadtxt("decoys_4states_rcc.csv", dtype="float", delimiter=",")
X_cath_domains_set = np.loadtxt('CATHFINAL.txt', dtype="float",usecols=(range(1,27)), delimiter=",")
```
